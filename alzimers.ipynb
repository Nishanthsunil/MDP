```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
data["Group"]=LE.fit_transform(data["Group"])
data["M/F"]=LE.fit_transform(data["M/F"])
data["Hand"]=LE.fit_transform(data["Hand"])
```

```python
x=data.drop(["Subject ID","MRI ID","Group",'Hand'],axis=1)
y=data.Group
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_lr)
```

```python
print(lr.score(x_test,y_test))
print(lr.score(x_train,y_train))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred_lr)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred_dtc=dtc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=11,random_state=0)
rfc.fit(x_train,y_train)
y_pred_rfc=rfc.predict(x_test)

```

```python
accuracy_score(y_test,y_pred_rfc)
```

```python
print(rfc.score(x_test,y_test))
print(rfc.score(x_train,y_train))
```

```python
from sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(lr,n_estimators=500,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7,random_state=0)
bg.fit(x_train,y_train)
y_pred_bg=bg.predict(x_test)
accuracy_score(y_test,y_pred_bg)
```

```python
print(bg.score(x_test,y_test))
print(bg.score(x_train,y_train))
```

```python
from sklearn.ensemble import BaggingClassifier
bg_rfc=BaggingClassifier(rfc,n_estimators=500,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7,random_state=0)
bg_rfc.fit(x_train,y_train)
y_pred_rfc=bg.predict(x_test)
accuracy_score(y_test,y_pred_rfc)
```
