```python
import pandas as pd
data=pd.read_csv("C:\\Users\\Nishanth S\\Downloads\\breast-cancer.csv")
print(data.to_string())
```

```python
data.isnull().sum()
```

```python
data.id.value_counts()
```

```python
import seaborn as sns
sns.heatmap(data.corr(),annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
data["diagnosis"]=LE.fit_transform(data["diagnosis"])
```

```python
data
```

```python
x=data.drop(["diagnosis","id"],axis=1)
print(x.to_string())
```

```python
x.shape
```

```python
y=data["diagnosis"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pred_lr=lr.predict(x_test)
accuracy_score(y_test,y_pred_lr)
```

```python
print(accuracy_score(y_train,lr.predict(x_train)))
print(accuracy_score(y_test,lr.predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
y_pred_dtc=dtc.predict(x_test)
accuracy_score(y_test,y_pred_dtc)
```

```python
print(accuracy_score(y_train,dtc.predict(x_train)))
print(accuracy_score(y_test,dtc.predict(x_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=0)
rfc.fit(x_train,y_train)
```

```python
y_pred_rfc=rfc.predict(x_test)
accuracy_score(y_test,y_pred_rfc)
```

```python
print(accuracy_score(y_train,rfc.predict(x_train)))
print(accuracy_score(y_test,rfc.predict(x_test)))
```

Logistic regression

```python
from sklearn.ensemble import BaggingClassifier
bg_lr=BaggingClassifier(lr,n_estimators=500,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7)
bg_lr.fit(x_train,y_train)
y_pred_bg_lr=bg_lr.predict(x_test)
accuracy_score(y_test,y_pred_bg_lr)
```

```python
print(accuracy_score(y_train,bg_lr.predict(x_train)))
print(accuracy_score(y_test,bg_lr.predict(x_test)))
```

```python
confusion_matrix(y_test,y_pred_bg_lr)
```

random forest

```python
from sklearn.ensemble import BaggingClassifier
bg_rfc=BaggingClassifier(rfc,n_estimators=500,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7)
bg_rfc.fit(x_train,y_train)
y_pred_bg_rfc=bg_rfc.predict(x_test)
accuracy_score(y_test,y_pred_bg_rfc)
```

```python
print(accuracy_score(y_train,bg_rfc.predict(x_train)))
print(accuracy_score(y_test,bg_rfc.predict(x_test)))
```

```python
confusion_matrix(y_test,y_pred_bg_rfc)
```
